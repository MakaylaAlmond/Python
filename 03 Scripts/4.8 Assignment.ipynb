# New   spending_flag   made

```python
#making  customer fequency column
ords_prods_mergedd['Customer_frequency'] = ords_prods_mergedd.groupby(['user_id'])['days_since_prior_order'].transform(np.max)
```

```python
ords_prods_mergedd.loc[ords_prods_mergedd['Customer_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_mergedd.loc[(ords_prods_mergedd['Customer_frequency'] > 10) & (ords_prods_mergedd['max_order'] <=20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_mergedd.loc[ords_prods_mergedd['Customer_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
